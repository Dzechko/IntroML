```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense


l0 = Dense(units=1, input_shape=[1])
# defining a neural network using Sequential(one line means one layer) layertype=Dense
#unit = 1 means we have one dense layer with one neuron
#we have to define shape of input data
l0 = Dense(units=1, input_shape=[1])
model = Sequential(l0)

#model guess possible relation and error and try to reduce the error
#loss calculate error and sgd is an optimizer which predict next guess
model.compile(optimizer='sgd', loss='mean_squared_error')

#numpy arrays to make process easy
xs = np.array([3.0, 4.0 , 2.0, 5.0, 9.0, 3.2, 5.8], dtype=float)
ys = np.array([8.0, 10.0 , 6.0, 12.0, 20.0, 8.4, 13.6], dtype=float)

model.fit(xs, ys, epochs=500)
print(model.predict([8]))

print("Here is what I learned: {}".format(l0.get_weights()))

```
